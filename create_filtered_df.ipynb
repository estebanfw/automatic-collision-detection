```python
import pandas as pd
import datetime as dt
```

```python
data=pd.read_pickle("dataframe/_PRUEBA_df_20211225_143242.pkl")
data.reset_index(inplace=True)
data.drop(['index'], inplace=True, axis=1)
print(data.shape)
data.head()
```

```python
# SELECTING DATA TO BUILD MODEL

print("Building dataframe... \n")
aux1=data[(data["TARGET_PC"]>=-6)]


# APPEND SUBPART OF DATAFRAMES EVENTS WITH PROBABILITIES LOWER THAN 10-6 TO CREATE AN EQUALLY DISTRIBUITED PROBABILITY DATAFRAME
print("Dataframe final size {} x {}".format(aux1.shape[0],aux1.shape[1]))
```

```python
# CREATE PICKLE FILE TO LOAD WHEN NEEDED
print("Creating pickle file... \n")
filename="./dataframe/_PRUEBA_df_filtered_{}.pkl".format(dt.datetime.now().strftime("%Y%m%d_%H%M%S"))
print("Saving dataframe for future usage filename = {}".format(filename))
aux1.to_pickle(filename)
```
